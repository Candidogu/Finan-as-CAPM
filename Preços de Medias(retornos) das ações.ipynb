```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import datetime as dt
```

```python
ativo = 'VALE3.SA'
data_inicio = "2015-01-01"
data_fim = "2024-11-16"

dados = yf.download(ativo, start=data_inicio, end=data_fim)

dados['Retorno'] = dados['Close'].pct_change()

preco = dados['Close'].squeeze() 

print('\nFechamento')
print(preco)

media_retornos = dados['Retorno'].mean()

plt.figure(figsize=(12, 5))
graf_linha = sns.lineplot(x=dados.index, y=preco, color = "b")
plt.xlabel('Data')
plt.ylabel('Preo')
graf_linha.set_title(f"Preo do {ativo}")
plt.grid(True)
plt.show

plt.figure(figsize=(12, 5))
sns.lineplot(x=dados.index, y=dados['Retorno'].squeeze()*100, color='b')
plt.xlabel('Data')
plt.ylabel('Media dos Retornos (%)')
plt.title(f'Media dos Retornos Diarios de {ativo}')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
